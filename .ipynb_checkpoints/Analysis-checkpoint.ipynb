```python
import numpy as np
import seaborn as sns
import pandas as pd
import pandas_bokeh
import plotly
import panel as pn
import plotly.express as px
pn.extension()
from panel import widgets
import matplotlib.pyplot as plt
```

```python
#high price data 
high_price_data= pd.read_csv('results.csv')
high_price_data
```

```python
pivot = high_price_data.pivot_table(index=['Hour'], values=['High Price'], aggfunc='mean')
pivot=pivot.reset_index()                             
pivot
```

```python
plt.plot(pivot['Hour'], pivot['High Price'], color='blue', marker='o')
plt.title('Average High Stock Price Per hour during May 11 2021 ', fontsize=14)
plt.xlabel('Hour (24hr)', fontsize=14)
plt.ylabel('Average High Price', fontsize=14)
plt.grid(True)
plt.show()
```

```python
df = high_price_data.pivot(index='Hour', columns='Ticker Name', values='High Price')
df
```

```python
plt.figure(figsize=(10,10))
sns.lineplot(data=df)
plt.title('High Price Per Stock on May 11 2021', fontsize=14)
plt.xlabel('Hour (24hr)', fontsize=14)
plt.ylabel('Stock High Price', fontsize=14)
```
